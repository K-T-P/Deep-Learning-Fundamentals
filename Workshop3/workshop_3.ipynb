```python
def accuracy(y_pred, y):
    """
    Calculating the accuracy of given batch.
    input(s):
    y_pred (ndarray): predicted logits or the output of the model.
    y (ndarray): labels
    """
    pred_idx = y_pred.max(1, keepdim=True)[1]
    correct = pred_idx.eq(y.view_as(pred_idx)).sum().item()
    return correct / len(y)
```

```python
# This cell is for working with learning rate scheduler and early stopping
# and is just for your test

history = dict()
history['train_loss']= list()
history['train_acc']= list()
history['val_loss']= list()
history['val_acc']= list()


class EarlyStopping:
    def __init__(self, patience=5, verbose=False):
        self.patience = patience
        self.verbose = verbose
        self.counter = 0
        self.best_loss = None
        self.early_stop = False

    def __call__(self, val_loss):
        if self.best_loss is None or val_loss < self.best_loss:
            self.best_loss = val_loss
            self.counter = 0
        else:
            self.counter += 1
            if self.counter >= self.patience:
                self.early_stop = True
                if self.verbose:
                    print("Early stopping")

    def should_stop(self):
        return self.early_stop


from torch.optim.lr_scheduler import StepLR
scheduler = StepLR(optimizer, step_size=10, gamma=0.1)

# Early stopping
early_stopping = EarlyStopping(patience=3, verbose=True)


for epoch in range(n_epochs):
    running_loss = 0
    running_acc = 0

    model.train()
    for idx, (X, y) in enumerate(tqdm(trainloader)):
        optimizer.zero_grad()
        X, y = X.to(device), y.to(device)
        y_hat = model(X)
        loss = criterion(y_hat, y)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        running_acc += accuracy(y_hat, y)
    else:
        running_loss = running_loss/ len(trainloader)
        running_acc = running_acc / len(trainloader)
        history['train_loss'].append(running_loss)
        history['train_acc'].append(running_acc)
        print(f'Epoch {epoch+1}/{n_epochs} : training loss: {round(running_loss,3)}, training acc: {round(running_acc*100,2)}')



    if epoch% val_per_epoch == val_per_epoch-1:
        running_loss = 0
        running_acc = 0
        model.eval()
        with torch.no_grad():
            for idx, (X, y) in enumerate(tqdm(valloader)):
                X, y = X.to(device), y.to(device)
                y_hat = model(X)
                loss = criterion(y_hat, y)

                running_loss += loss.item()
                running_acc += accuracy(y_hat, y)
            else:
                running_loss = running_loss/ len(valloader)
                running_acc = running_acc / len(valloader)
                history['val_loss'].append(running_loss)
                history['val_acc'].append(running_acc)
                print(f'Epoch {epoch+1}/{n_epochs} : validation loss: {round(running_loss,3)}, validation acc: {round(running_acc*100,2)}')



    # Check for early stopping
    early_stopping(running_loss)
    if early_stopping.should_stop():
      print("Early stopping triggered")
      break

    # Step the scheduler
    scheduler.step()
```

```python
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
from torch import nn, optim
from torch.utils.data import DataLoader, Dataset
from tqdm import tqdm
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score, accuracy_score
```

```python
cancer_x_resampled = pd.read_csv('cancer_x_resampled.csv').drop(columns='Unnamed: 0').to_numpy(dtype=np.float32)
cancer_y_resampled = pd.read_csv('cancer_y_resampled.csv').drop(columns='Unnamed: 0').to_numpy(dtype=np.float32)
cancer_x_test = pd.read_csv('cancer_x_test.csv').drop(columns='Unnamed: 0').to_numpy(dtype=np.float32)
cancer_y_test = pd.read_csv('cancer_y_test.csv').drop(columns='Unnamed: 0').to_numpy(dtype=np.float32)
```

```python
class TrainDataSet(Dataset):
    def __init__(self, X_train, Y_train, transform=None):
        self.length = len(X_train)
        self.x_train = X_train
        self.y_train = Y_train
        self.transform = transform

    def __len__(self):
        return self.length

    def __getitem__(self, idx):
        item_x = torch.tensor(self.x_train[idx])
        item_y = torch.tensor(self.y_train[idx])
        return item_x, item_y
```

```python
class TestDataSet(Dataset):
    def __init__(self, X_train, Y_train, transform=None):
        self.length = len(X_train)
        self.x_train = X_train
        self.y_train = Y_train
        self.transform = transform

    def __len__(self):
        return self.length

    def __getitem__(self, idx):
        item_x = torch.tensor(self.x_train[idx])
        item_y = torch.tensor(self.y_train[idx])
        return item_x, item_y
```

```python
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

transform = transforms.Compose([
    transforms.ToTensor(),
])
```

```python
train_dataset = TrainDataSet(cancer_x_resampled, cancer_y_resampled,transform=transform)
test_dataset = TestDataSet(cancer_x_test,cancer_y_test,transform=transform)

train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=True)
```

```python
# @title Default title text
class MLP(nn.Module):
    def __init__(self, input_dim, hidden_dim1, hidden_dim2, output_dim, dropout_prob=0.5, device='cpu'):
        super(MLP, self).__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(in_features=input_dim, out_features= hidden_dim1, device=device, dtype=torch.float32)
        self.fc2 = nn.Linear(in_features=hidden_dim1, out_features= hidden_dim2, device=device, dtype=torch.float32)
        self.fc3 = nn.Linear(in_features=hidden_dim2, out_features=output_dim, device=device, dtype=torch.float32)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.relu(x)
        x = self.fc3(x)
        x = self.sigmoid(x)
        return x
```

```python
model = MLP(cancer_x_resampled.shape[1], 4, 2, 1)
model.to(device)

lr = 0.1
optimizer = optim.SGD(model.parameters(), lr = lr)
loss_fn = nn.BCELoss()
```

```python
from torchsummary import summary

summary(model, (1,36))
```

```python
train_loss = []
train_acc = []
test_acc = []
test_loss = []
test_predictions = []
test_true_labels = []

epochs = 40

for epoch in tqdm(range(epochs)):
    model.train()
    train_size = 0
    running_loss = 0
    running_acc = 0
    for _, x in enumerate(train_dataloader):
        input, label = x
        output = model(input.to(device))
        optimizer.zero_grad()
        loss = loss_fn(output, label)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
        running_acc += torch.sum(torch.eq(output.detach().round(), label)).item()
        train_size += len(label)
    else:
        running_acc /= train_size
        train_loss.append(running_loss)
        train_acc.append(running_acc * 100)

    with torch.inference_mode():
        val_size = 0
        val_acc = 0
        val_loss = 0
        for _, x in enumerate(test_dataloader):
            input, label = x
            output = model(input)
            test_predictions.extend(output.round().squeeze().tolist())
            test_true_labels.extend(label.squeeze().tolist())
            val_acc += torch.eq(output.round(), label).sum().item()
            val_loss += loss_fn(output, label)
            val_size += len(label)
        val_acc /= val_size
        test_acc.append(val_acc * 100)
        test_loss.append(val_loss)

        if (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch + 1}, Train_Acc : {running_acc:.2f}, Train_loss : {running_loss:.2f}, Test_Acc : {val_acc:.2f}, Test_loss : {val_loss:.2f}")
```

```python
test_confusion_matrix = confusion_matrix(y_true = test_true_labels, y_pred = test_predictions)
test_f1_score = f1_score(y_true = test_true_labels , y_pred = test_predictions)
test_precision_score = precision_score(y_true = test_true_labels, y_pred = test_predictions)
test_recall_score = recall_score(y_true = test_true_labels, y_pred = test_predictions)
test_accuracy_score = accuracy_score(y_true = test_true_labels, y_pred = test_predictions)

print(f"Test Confusion matrix :\n{test_confusion_matrix}")
print(f"Test F1 Score : {test_f1_score}")
print(f"Test Precision Score : {test_f1_score}")
print(f"Test Recall Score : {test_recall_score}")
print(f"Test Accuracy Score : {test_accuracy_score}")
```
